```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
#Create a column called age_group
df["age_group"]=pd.qcut(df["age"],q=4,labels=["Young Adult","Adult","Middle-aged","Senior"])
```

```python
df[["age","age_group"]].head(5)
```

```python
#Create column purchased_frequency_days
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchased_frequency_days"]=df["frequency_of_purchases"].replace(frequency_mapping)
```

```python
df[["frequency_of_purchases","purchased_frequency_days"]].head()
```
